```python
all_sentences = all_builder(data, [['race', 'White'], ['race', 'Black'], ['race', 'Hispanic'], ['race', 'Asian'], ['sex', 'Male'], ['sex', 'Female'], ['criminal_history', 'No Prior Criminal Convictions'], ['criminal_history', '2 Prior Criminal Convictions'], ['criminal_history', '3 Prior Criminal Convictions']])

for key, value in all_sentences.items():
    
    print(key, "average sentence length = ", np.mean(value))
```
